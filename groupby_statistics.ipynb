```python
import pandas as pd
import scipy.stats as st

# build a dataframe
d = {
    'one':[1,1,2,2,2],
    'two':[2,4,6,6,10]
}

df = pd.DataFrame(d)
df = df.set_index('one')
df
```

```python
# group by the 'one' colume
gdf = df.groupby('one')
```

```python
def mymode(d):
    return(st.mode(d).mode[0][0])
    
ndf = pd.DataFrame(gdf.count())
ndf.rename(columns={'two':'count'}, inplace=True)
ndf['min'] = gdf.min()
ndf['mean'] = gdf.mean()
ndf['max'] = gdf.max()
ndf['median'] = gdf.median()
ndf['mode'] = gdf.apply(mymode)
ndf['var'] = gdf.var()
ndf['std'] = gdf.std()
ndf['sem'] = gdf.sem()
ndf
```
